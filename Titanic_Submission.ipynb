```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
test_data=pd.read_csv("test.csv")
```

```python
total_data=[train_data,test_data]
#it is a list so we cant operation on it directly
```

```python
for dataset in total_data:
    #dataset['title'] = dataset['Name'].str.extract(' ([A-Za-z] +) \ .',expand=False)
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train_data['Title'].value_counts()
```

```python
test_data['Title'].value_counts()
```

```python
title_map = {'Mr':0,'Miss':1,'Mrs':2,'Master':3,'Dr':3,'Rev':3,
             'Col':3,'Mlle':3,'Major':3,'Capt':3,'Lady':3,'Mme':3,'Sir':3,'Ms':3,'Jonkheer':3,'Countess':3,'Don':3,'Dona':3}
```

```python
for dataset in total_data:
    dataset['Title'] = dataset['Title'].map(title_map)
```

```python
train_data.head(6)
```

```python
train_data=train_data.drop(['Name'],axis=1)
```

```python
test_data=test_data.drop(['Name'],axis=1)
```

```python
total_data=[train_data,test_data]
```

```python
for dataset in total_data:
    dataset['Sex'] = dataset['Sex'].map({'female':1,'male':0}).astype(int)
```

```python
train_data['Age'].fillna(train_data.groupby('Title')['Age'].transform('median'),inplace=True)
test_data['Age'].fillna(test_data.groupby('Title')['Age'].transform('median'),inplace=True)
```

```python
train_data.head(4)
```

```python
test_data.head(4)
```

```python
for dataset in total_data:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_data.head()
```

```python
comm_embark = train_data['Embarked'].dropna().mode()[0]
total_data = [train_data,test_data]
```

```python
for dataset in total_data:
    dataset['Embarked'] = dataset['Embarked'].fillna(comm_embark)
```

```python
for dataset in total_data:
    dataset['Embarked'] = dataset['Embarked'].map({'S':0,'Q':1,'C':2})
    
```

```python
train_data.head(4)
```

```python
train_data['Fare'].fillna(train_data.groupby('Pclass')['Fare'].transform('median'),inplace=True)
test_data['Fare'].fillna(test_data.groupby('Pclass')['Fare'].transform('median'),inplace=True)
train_data.head(4)
```

```python
for dataset in total_data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
train_data.head(4)
```

```python
for dataset in total_data:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
```

```python
train_data.head(5)
```

```python
train_data=train_data.drop(['SibSp','Parch','Ticket','Cabin'],axis=1)
```

```python
test_data=test_data.drop(['SibSp','Parch','Ticket','Cabin'],axis=1)
```

```python
train_data = train_data.drop(['PassengerId'],axis=1)
```

```python
train_data.info()
train_data.isnull().sum()
```

```python
test_data.info()
test_data.isnull().sum()
test_data['Title']
```

```python
X_train = train_data.drop("Survived", axis=1)
Y_train = train_data["Survived"]
X_test  = test_data.drop("PassengerId", axis=1).copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
logreg =LogisticRegression()
logreg.fit(X_train,Y_train)
Y_pred = logreg.predict(X_test)
acc = round(logreg.score(X_train,Y_train)*100,2)
acc
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
submission = pd.DataFrame({
        "PassengerId": test_data["PassengerId"],
        "Survived": Y_pred
    })
```

```python
submission.to_csv('submission.csv',index=False)
```
